```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('movies_dataset.json')
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.gross
```

## Data Preparation

```python
df.drop(axis = 1, columns = ['movie_imdb_link', 'num_critic_for_reviews', 'genres'], inplace = True)
df.head()
```

```python
df.shape
```

```python
df.title_year = df.title_year.astype('Int64')
```

```python
df.head(5)
```

```python
rename_dict = {'gross' : 'movie_income', 'language' : 'Language', 'budget' : 'movie_budget'}
df.rename(columns = rename_dict, inplace = True)
df.head(5)
```

## Data Analysis and Visualization

```python
ax = df.duration.hist(bins=50, figsize = (8,6), grid = False)
ax.set_xlabel('Movie Duration')
ax.set_ylabel('Count')
ax.set_title("Movie Duration Histogram")
plt.show()
```

```python
ax = df.imdb_score.hist(bins=30, figsize = (8,6), grid = False)
ax.set_xlabel('Score')
ax.set_ylabel('Count')
ax.set_title("Movies IMDB Score Histogram")
plt.show()
```

```python
low_score = df[df.imdb_score < 4]
low_score.head()
low_score.shape
```

```python
df[df.country == 'USA'].imdb_score.hist(bins=25, figsize = (8,6), grid = False)
plt.show()
```

```python
ax = df[df.country == 'France'].imdb_score.hist(bins=15, figsize = (8,6), grid = False)
ax.set_xlabel('Score')
ax.set_ylabel('Count')
ax.set_title("French Movies IMDB Histogram Score")
plt.show()
```

```python
count_per_year = df.groupby('title_year').size()
count_per_year.head()
```

```python
ax = count_per_year.plot(figsize = (8,6))
ax.set_xlabel('Years')
ax.set_ylabel('Number of Movies')
ax.set_title('Number of Movies Evaluation')
plt.show()
```

```python
avg_score = df.groupby('title_year')['imdb_score'].mean()
avg_score
```

```python
ax = avg_score.plot.bar(figsize = (25,10), title = 'Avg IMDB Score Per year' )
ax.set_ylabel('Avg Score')
```

```python
ax = df.groupby('title_year')['movie_budget'].sum().plot()
ax.set_xlabel('Years')
ax.set_ylabel('Budget ')
ax.set_title('Budget per Year')
plt.show()
```

```python
df_corr = df[['imdb_score', 'movie_income']]
df_corr
```
