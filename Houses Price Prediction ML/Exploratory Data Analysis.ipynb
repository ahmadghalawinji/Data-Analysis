### Missing Values


```python
# Create a list of features with more than one missing value
null_features = [features for features in dataset.columns if dataset[features].isnull().sum() > 1]

# Iterate over the null features and print the percentage of missing values for each
for feature in null_features:
    print(feature, np.round(dataset[feature].isnull().mean(), 4), ' % missing values')
```

```python
# Iterate over the null features
for feature in null_features:
    # Create a copy of the dataset to avoid modifying the original data
    data = dataset.copy()
    
    # Create a new binary variable to indicate if the observation was missing or not
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # Group the data by the binary variable and plot the median SalePrice for each group
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
    # Add a title to the plot indicating the feature being analyzed
    plt.title(feature)
    
    # Display the plot
    plt.show()
```

### Numerical Variables

### Temporal Variables(i.e. Datetime variables)


```python
# Create a list of numerical features that contain the strings 'Yr' or 'Year'
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
# Group the dataset by the year the house was sold and plot the median sale price for each year
dataset.groupby('YrSold')['SalePrice'].median().plot()

# Add a label to the x-axis indicating the year the house was sold
plt.xlabel('Year Sold')

# Add a label to the y-axis indicating the median house price
plt.ylabel('Median House Price')

# Add a title to the plot
plt.title("House Price vs Year Sold")
plt.show()
```

#### This info doesn't look very promising as with the years passing the sales price is dropping so we will try to find more info.



```python
# Iterate over each feature in the year_feature list
for feature in year_feature:
    # Check if the feature is not 'YrSold'
    if feature != 'YrSold':
        # Create a copy of the dataset
        data = dataset.copy()
        # Calculate the difference between the year variable and the year the house was sold
        data[feature] = data['YrSold'] - data[feature]
        
        # Create a scatter plot of the feature vs. sale price
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
    
```

### Continuous Variables

```python
# Create a list of continuous features by iterating over the numerical_features list and excluding discrete, year, and Id features
continuous_features = [feature for feature in numerical_features if feature not in discrete_features+year_feature+['Id']]

# Print the count of continuous features
print("Continuous Feature Count {}".format(len(continuous_features)))
```

```python
# Iterate over each feature in the continuous_features list
for feature in continuous_features:
    # Create a copy of the dataset
    data = dataset.copy()
    # Plot a histogram of the current feature using the hist() method with 50 bins
    data[feature].hist(bins=50)
    
    # Add a label to the x-axis indicating the feature
    plt.xlabel(feature)
    # Add a label to the y-axis indicating the count of data points
    plt.ylabel('Count')
    # Add a title to the plot indicating the feature
    plt.title(feature)
    # Display the plot
    plt.show()
```

```python
# we will now use logarithmic transformation
for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalesPrice')
        plt.title(feature)
        plt.show()
```

### Outliers

```python
for feature in continuous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
