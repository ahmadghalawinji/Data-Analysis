## Analyzing Spotify Dataset Using Seaborn, Matplotlib, and Pandas to Get Valuable Insights

In this project, we will be analyzing a Spotify dataset using Python libraries such as Seaborn, Matplotlib, and Pandas. Our goal is to gain valuable insights into the data and understand how different features of songs are related to each other.
Dataset

The dataset we will be using is the Spotify Dataset 1921-2020, which contains over 160,000 songs and their attributes such as tempo, duration, danceability, energy, and many others. The dataset can be found here.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tracks.csv')
df.head()
```

```python
df['duration'] = df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms', inplace=True, axis=1)
```

```python
df.head(10)
```

## Correlation Map

```python
corr = df.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14, 16))
heatmap = sns.heatmap(corr, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', annot=True, linewidth=1, linecolor='Black')
heatmap.set_title('Correlation Heatmap')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
sample_df = df.sample(int(0.0005 * len(df)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'b').set(title = 'Correlation between Loudness and Energy')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'c').set(title = 'Correlation between Pupularity and Accousticness')
plt.show()
```

```python
df = df.set_index("release_date")
```

```python
df['dates'] = df.index.get_level_values('release_date')
df.dates = pd.to_datetime(df.dates)
years = df.dates.dt.year
df
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of Songs per Year')
plt.show()
```

```python
total_duration = df.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims) # create a Figure object first
fig = sns.barplot(x = years, y =total_duration, ax = ax, errwidth = False).set(title = 'Correlation between Year and Duration')
plt.xticks(rotation = 60)
plt.show()
```

```python
total_dr = df.duration
sns.set_style(style = 'whitegrid')
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize = fig_dims) # create a Figure object first
sns.lineplot(x=years, y=total_dr, ax=ax)
plt.xticks(rotation = 60)
```

# Features Dataset

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.head(5)
```

```python
plt.title('Duration of Songs for Different Genres')
sns.color_palette('rocket', as_cmap = True)
sns.barplot( y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel('Duration in milli second')
plt.ylabel('Genres')
plt.show()
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10, 5))
famous = df_genre.sort_values('popularity', ascending = False).head(20)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = 'Top Genres by Popularity')
plt.show()
```
