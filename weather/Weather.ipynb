Welcome to our weather data analysis project, where we will be using the seaborn library to analyze a dataset of weather information. This project is designed to showcase how data analysis can be used to gain insights and make informed decisions based on data. Our goal is to extract meaningful information from the weather dataset and present it in a clear and concise manner.

## About the Dataset

The weather dataset we will be using contains information about various weather conditions such as temperature, humidity, wind speed, etc. We will be analyzing this data to identify patterns, trends, and relationships between different variables.

## About Seaborn Library

Seaborn is a Python data visualization library that is based on the popular Matplotlib library. It provides a high-level interface for creating informative and attractive statistical graphics. Seaborn is particularly well-suited for exploratory data analysis and can help us quickly visualize relationships between variables in our dataset.

## Project Goals

Our primary goal for this project is to analyze the weather dataset and gain insights into the various weather conditions. We will be using various seaborn visualizations such as scatter plots, line plots, bar plots, and heatmaps to visualize the data and identify patterns. Additionally, we will be conducting statistical analyses to test hypotheses and make inferences about the data.


```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
df = pd.read_csv('Test.csv')
df.head()
```

```python
df.info()
```

```python
sns.barplot(x = df['humidity'].head(30), y = df['temperature'].head(30))
```

```python
sns.histplot(df['temperature'])
```

```python
sns.histplot(df['wind_speed'])
```

```python
sns.jointplot(x = df['humidity'], y = df['temperature'], kind='hex')
```

```python
sns.jointplot(x = df['humidity'], y = df['air_pollution_index'], kind='kde')
```

```python
sns.pairplot(df[['humidity','air_pollution_index','temperature','visibility_in_miles']])
```

```python
sns.stripplot(x = df['weather_type'], y = df['temperature'])
```

```python
sns.countplot(x = df['weather_type'], data=df)
```

```python
sns.lmplot(x = 'humidity', y = 'temperature', data = df)
```

```python
sns.lmplot(x = 'humidity', y = 'air_pollution_index', data = df)
```
